```python
# initialization
from rayoptics.environment import *

from rayoptics.optical import doe
from rayoptics.util.spectral_lines import spectra
```

## Compute a thin achromat

```python
def achromat(f, Va, Vb):
    phi_a = Va/(f*(Va - Vb))
    phi_b = Vb/(f*(Vb - Va))
    return phi_a, phi_b
```

# Create a new model

```python
opm = OpticalModel()
sm  = opm.seq_model
osp = opm.optical_spec
pm = opm.parax_model
```

## Define first order aperture and field for system

```python
osp.pupil = PupilSpec(osp, pupil_type=PupilType.EPD, value=5.0)
osp.field_of_view = FieldSpec(osp, field_type=FieldType.OBJ_HT, flds=[0., 1.0])
osp.spectral_region = WvlSpec([(486.1327, 0.5), (587.5618, 1.0), (656.2725, 0.5)], ref_wl=1)
```

## Define interface and gap data for the sequential model

```python
opm.radius_mode = False

sm.gaps[0].thi=1.0e10

sm.add_surface([0.0387, 1.0, 'N-BK7', 'Schott'])
s1 = sm.ifcs[sm.cur_surface]

sm.add_surface([0, 49.])
s2 = sm.ifcs[sm.cur_surface]
g2 = sm.gaps[sm.cur_surface]
```

```python
def refocus(opm):
    osp = opm.optical_spec
    pupil = [0., 1.]
    fi = 0
    wl = osp.spectral_region.reference_wvl
    fld, wvl, foc = osp.lookup_fld_wvl_focus(fi, wl)
    ray, ray_op, wvl = trace_base(opm, pupil, fld, wvl)
    ray = [RaySeg(*rs) for rs in ray]
    defoc = -ray[-1].p[1]/(ray[-2].d[1]/ray[-2].d[2])
    osp.defocus.focus_shift = defoc
    return defoc
```

```python
pupil = [0., 1.]
fi = 0
wl = osp.spectral_region.reference_wvl
fld, wvl, foc = osp.lookup_fld_wvl_focus(fi, wl)
ray, ray_op, wvl = trace_base(opm, pupil, fld, wvl)
ray = [RaySeg(*rs) for rs in ray]
list_ray(ray)
```

## Compute hybrid achromat

```python
V_doe = spectra['d']/(spectra['F'] - spectra['C'])
medium1 = sm.gaps[1].medium
nd = medium1.rindex(spectra['d'])
nC = medium1.rindex(spectra['C'])
nF = medium1.rindex(spectra['F'])
V_lens = (nd - 1)/(nF - nC)
f = 50.
phi_doe, phi_lens = achromat(f, V_doe, V_lens)
f, V_doe, V_lens, phi_doe, phi_lens
```

```python
c = phi_lens/(nd - 1);c
```

```python
s1.phase_element = doe.DiffractiveElement(ref_wl=spectra['d'], coefficients=[-phi_doe], phase_fct=doe.radial_phase_fct)
s1.profile.c = phi_lens/(nd - 1)
g2.thi = 49.
```

```python
opm.update_model()
```

```python
ray, ray_op, wvl = trace_base(opm, pupil, fld, wvl)
ray = [RaySeg(*rs) for rs in ray]
list_ray(ray)
```

## examine DOE phase calculation

```python
pupil = [0., 1.]
fi = 0
wl = osp.spectral_region.reference_wvl
fld, wvl, foc = osp.lookup_fld_wvl_focus(fi, wl)
ray_d, ray_op_d, wvl_d, opd_d = trace_with_opd(opm, pupil, fld, spectra['d'], foc)
ray_C, ray_op_C, wvl_C, opd_C = trace_with_opd(opm, pupil, fld, spectra['C'], foc)
ray_F, ray_op_F, wvl_F, opd_F = trace_with_opd(opm, pupil, fld, spectra['F'], foc)
```

```python
n_in_d = sm.rndx[wvl_d].iloc[0]
n_out_d = sm.rndx[wvl_d].iloc[1]
_, dW_d = rt.phase(sm.ifcs[1], ray_d[1][0], ray_d[0][1], ray_d[1][3], wvl_d, n_in_d, n_out_d)

n_in_C = sm.rndx[wvl_C].iloc[0]
n_out_C = sm.rndx[wvl_C].iloc[1]
_, dW_C = rt.phase(sm.ifcs[1], ray_C[1][0], ray_C[0][1], ray_C[1][3], wvl_C, n_in_C, n_out_C)

n_in_F = sm.rndx[wvl_F].iloc[0]
n_out_F = sm.rndx[wvl_F].iloc[1]
_, dW_F = rt.phase(sm.ifcs[1], ray_F[1][0], ray_F[0][1], ray_F[1][3], wvl_F, n_in_F, n_out_F)
```
